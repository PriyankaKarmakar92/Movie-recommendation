```python
 movies = pd.DataFrame({
    'Movie_Crew':
        [['Steven Spielberg', 'John Williams'],
        'Martin Scorsese, Robert De Niro',
        'Quentin Tarantino' ]
})

def fetch_director(crew):
    if isinstance(crew, list):
        return crew[0] if len(crew) > 0 else None
    elif isinstance(crew, str):
        return crew.split(',')[0].strip() if len(crew) > 0 else None
    else:
        return None
if 'Movie_Crew' in movies.columns:
    movies['Movie_Crew'] = movies['Movie_Crew'].apply(fetch_director)
else:
    print("Column 'Movie_Crew' not found in DataFrame.")

print(movies)
```
